```python
import pandas as pd
import numpy as np

print("Preprocessing started ")
```

```python
ratings = pd.read_csv("../data/raw/ratings.csv")
movies = pd.read_csv("../data/raw/movies.csv")
tags = pd.read_csv("../data/raw/tags.csv")
```

```python
ratings["timestamp"] = pd.to_datetime(ratings["timestamp"], unit="s")
```

```python
ratings = ratings.sort_values(["userId", "timestamp"])
ratings.head
```

```python
train_rows = []
test_rows = []

for user_id, user_df in ratings.groupby("userId"):
    train_rows.append(user_df.iloc[:-1])
    test_rows.append(user_df.iloc[-1:])
    
train_data = pd.concat(train_rows)
test_data = pd.concat(test_rows)
```

```python
print(train_data.columns)
print(test_data.columns)
```

```python
movies["genres"] = movies["genres"].str.lower().str.replace("|", " ")
```

```python
tags["tag"] = tags["tag"].str.lower().str.strip()
```

```python
import os

os.makedirs("../data/processed", exist_ok=True)
```

```python
train_data.to_csv("../data/processed/train_ratings.csv", index=False)
test_data.to_csv("../data/processed/test_ratings.csv", index=False)
movies.to_csv("../data/processed/movies_clean.csv", index=False)
tags.to_csv("../data/processed/tags_clean.csv", index=False)

print("Processed data saved ")
```

A time-aware trainâ€“test split was used (leave-last-one per user) to simulate real-world recommendation scenarios and prevent temporal data leakage.
